```python
import pandas as pd
from IPython.display import display

fon_data = [
    ["TI1", "Fon - Para Piyasasi", 243150, "TL"],
    ["IST", "Fon - Kısa Vadeli Borç", 83339, "TL"],
    ["IPV", "Fon - Eurobond", 40695, "TL"],
    ["TTA", "Fon - Altın", 30862, "TL"],
    ["AFT", "Fon - Teknoloji", 13289, "TL"],
    ["TMG", "Fon - Yabancı Hisse", 10406, "TL"],
    ["NST", "Fon - İnşaat", 9247, "TL"],
]

df_fon = pd.DataFrame(fon_data, columns=["Kod", "kategori", "Tutar_TL", "Para Birimi"])

maden_data = [
    ["Altin", "Kiymetli Maden", 15, "gr"],
    ["Gumus", "Kiymetli Maden", 465, "gr"],
]

df_maden = pd.DataFrame(maden_data, columns=["Varlik", "Kategori", "Miktar", "Birim"])

doviz_data = [
    ["EUR", "Döviz", 800],
    ["USD", "Döviz", 500],
]

df_doviz = pd.DataFrame(doviz_data, columns=["Varlık", "Kategori", "Miktar"])

print("==Fonlar==")
display(df_fon)

print("\n==Kiymetli Madenler==")
display(df_maden)

print("\n==Doviz==")
display(df_doviz)

print("\nFonlar toplam (TL):", df_fon["Tutar_TL"].sum())
```

```python
import pandas as pd
import numpy as np

# Fiyatlar
ALTIN_TL_GR = 6000.0
GUMUS_TL_GR = 98.0
USDTRY = 42.9372   # sende neyse
EURTRY = 50.4614   # sende neyse

def norm_cols(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out.columns = [c.strip().lower().replace(" ", "_") for c in out.columns]
    return out

df_fon_n = norm_cols(df_fon)
df_maden_n = norm_cols(df_maden)
df_doviz_n = norm_cols(df_doviz)

# --- FON ---
df_fon_port = df_fon_n.copy()
df_fon_port["tur"] = "fon"
df_fon_port["varlik"] = df_fon_port["kod"]
df_fon_port["alt_kategori"] = df_fon_port["kategori"]
df_fon_port["miktar"] = np.nan
df_fon_port["birim"] = "TL"
df_fon_port["birim_fiyat_tl"] = np.nan
df_fon_port["deger_tl"] = df_fon_port["tutar_tl"]
df_fon_port = df_fon_port[["tur","varlik","alt_kategori","miktar","birim","birim_fiyat_tl","deger_tl"]]

# --- MADEN ---
df_maden_port = df_maden_n.copy()
df_maden_port["tur"] = "maden"
df_maden_port["varlik_norm"] = (
    df_maden_port["varlik"].astype(str).str.lower().str.strip().str.replace("ı","i")
)
df_maden_port["birim_fiyat_tl"] = df_maden_port["varlik_norm"].map({
    "altin": ALTIN_TL_GR,
    "gumus": GUMUS_TL_GR
})
df_maden_port["alt_kategori"] = df_maden_port["varlik"]
df_maden_port["deger_tl"] = df_maden_port["miktar"] * df_maden_port["birim_fiyat_tl"]
df_maden_port = df_maden_port[["tur","varlik","alt_kategori","miktar","birim","birim_fiyat_tl","deger_tl"]]

# --- DOVIZ ---
df_doviz_port = df_doviz_n.copy()
df_doviz_port["tur"] = "doviz"
var_col = "varlık" if "varlık" in df_doviz_port.columns else "varlik"

d_var = df_doviz_port[var_col].astype(str).str.lower().str.strip().str.replace("ı","i")
df_doviz_port["birim"] = df_doviz_port[var_col]
df_doviz_port["birim_fiyat_tl"] = d_var.map({"usd": USDTRY, "eur": EURTRY})
df_doviz_port["deger_tl"] = df_doviz_port["miktar"] * df_doviz_port["birim_fiyat_tl"]
df_doviz_port["varlik"] = df_doviz_port[var_col]
df_doviz_port["alt_kategori"] = df_doviz_port[var_col]
df_doviz_port = df_doviz_port[["tur","varlik","alt_kategori","miktar","birim","birim_fiyat_tl","deger_tl"]]

# --- BİRLEŞTİR ---
portfoy = pd.concat([df_fon_port, df_maden_port, df_doviz_port], ignore_index=True)
portfoy["deger_tl"] = pd.to_numeric(portfoy["deger_tl"], errors="coerce")

toplam = portfoy["deger_tl"].sum()
portfoy["agirlik_%"] = (portfoy["deger_tl"] / toplam * 100).round(2)

print("Toplam Portfoy (TL):", round(toplam, 2))
portfoy.sort_values("deger_tl", ascending=False)
```

```python
import pandas as pd
import numpy as np

# --- Güncel fiyat/kurlar (senin kullandığın değerleri koru ya da güncelle) ---
ALTIN_TL_GR = 6000.0
GUMUS_TL_GR = 98.0
USDTRY = 42.9372
EURTRY = 50.4614

# --- Fon adetleri ---
fon_adet = {
    "TI1": 184,
    "IST": 553,
    "IPV": 596,
    "TTA": 50819,
    "AFT": 15689,
    "TMG": 9000,
    "NST": 8970,
    "BTE": 1848,
    "IJB": 826
}

def norm_cols(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out.columns = [c.strip().lower().replace(" ", "_") for c in out.columns]
    return out

# --- Mevcut tabloları normalize ---
df_fon_n = norm_cols(df_fon)
df_maden_n = norm_cols(df_maden)
df_doviz_n = norm_cols(df_doviz)

# --- BTE & IJB'yi df_fon'a ekle (TL değerleri verilmiş) ---
ek_fonlar = pd.DataFrame(
    [
        ["BTE", "Fon - Oyun & Teknoloji", 7181, "TL"],
        ["IJB", "Fon - Dijital Oyun", 4880, "TL"],
    ],
    columns=["kod", "kategori", "tutar_tl", "para_birimi"]
)

# para_birimi kolonu sende yoksa sorun olmasın diye ekleyelim
if "para_birimi" not in df_fon_n.columns:
    df_fon_n["para_birimi"] = "TL"

df_fon_n = pd.concat([df_fon_n, ek_fonlar], ignore_index=True)

# --- Fonlara adet + birim fiyat ekle ---
df_fon_n["adet"] = df_fon_n["kod"].map(fon_adet)
df_fon_n["birim_fiyat_tl"] = (df_fon_n["tutar_tl"] / df_fon_n["adet"]).round(4)

# --- Fon portföy tablosu ---
df_fon_port = df_fon_n.copy()
df_fon_port["tur"] = "fon"
df_fon_port["varlik"] = df_fon_port["kod"]
df_fon_port["alt_kategori"] = df_fon_port["kategori"]
df_fon_port["miktar"] = df_fon_port["adet"]
df_fon_port["birim"] = "adet"
df_fon_port["deger_tl"] = df_fon_port["tutar_tl"]

df_fon_port = df_fon_port[["tur","varlik","alt_kategori","miktar","birim","birim_fiyat_tl","deger_tl"]]

# --- Maden portföy tablosu ---
df_maden_port = df_maden_n.copy()
df_maden_port["tur"] = "maden"
df_maden_port["varlik_norm"] = (
    df_maden_port["varlik"].astype(str).str.lower().str.strip().str.replace("ı","i")
)
df_maden_port["birim_fiyat_tl"] = df_maden_port["varlik_norm"].map({
    "altin": ALTIN_TL_GR,
    "gumus": GUMUS_TL_GR
})
df_maden_port["alt_kategori"] = df_maden_port["varlik"]
df_maden_port["deger_tl"] = df_maden_port["miktar"] * df_maden_port["birim_fiyat_tl"]
df_maden_port = df_maden_port[["tur","varlik","alt_kategori","miktar","birim","birim_fiyat_tl","deger_tl"]]

# --- Döviz portföy tablosu ---
df_doviz_port = df_doviz_n.copy()
df_doviz_port["tur"] = "doviz"
var_col = "varlık" if "varlık" in df_doviz_port.columns else "varlik"
d_var = df_doviz_port[var_col].astype(str).str.lower().str.strip().str.replace("ı","i")

df_doviz_port["birim"] = df_doviz_port[var_col]
df_doviz_port["birim_fiyat_tl"] = d_var.map({"usd": USDTRY, "eur": EURTRY})
df_doviz_port["deger_tl"] = df_doviz_port["miktar"] * df_doviz_port["birim_fiyat_tl"]
df_doviz_port["varlik"] = df_doviz_port[var_col]
df_doviz_port["alt_kategori"] = df_doviz_port[var_col]
df_doviz_port = df_doviz_port[["tur","varlik","alt_kategori","miktar","birim","birim_fiyat_tl","deger_tl"]]

# --- Birleştir ---
portfoy = pd.concat([df_fon_port, df_maden_port, df_doviz_port], ignore_index=True)
portfoy["deger_tl"] = pd.to_numeric(portfoy["deger_tl"], errors="coerce")

toplam = portfoy["deger_tl"].sum()
portfoy["agirlik_%"] = (portfoy["deger_tl"] / toplam * 100).round(2)

print("Toplam Portfoy (TL):", round(toplam, 2))

# --- Kurumsal özet ---
portfoy_sorted = portfoy.sort_values("deger_tl", ascending=False)
display(portfoy_sorted)

dagilim = portfoy.groupby("tur")["deger_tl"].sum().sort_values(ascending=False)
dagilim_yuzde = (dagilim / toplam * 100).round(2)

print("\n--- Tur Bazli Dagilim (TL) ---")
print(dagilim)

print("\n--- Tur Bazli Dagilim (%) ---")
print(dagilim_yuzde)

ilk3 = portfoy_sorted.head(3)["deger_tl"].sum()
ilk5 = portfoy_sorted.head(5)["deger_tl"].sum()
print("\nIlk 3 varlik agirlik %:", round(ilk3/toplam*100, 2))
print("Ilk 5 varlik agirlik %:", round(ilk5/toplam*100, 2))
```

```python
def senaryo(df, degisimler):
    df2 = df.copy()
    for varlik, oran in degisimler.items():
        mask = df2["varlik"] == varlik
        df2.loc[mask, "deger_tl"] *= (1 + oran)
    toplam = df2["deger_tl"].sum()
    df2["agirlik_%"] = (df2["deger_tl"] / toplam * 100).round(2)
    return df2, toplam

# ornek senaryo
yeni_portfoy, yeni_toplam = senaryo(
    portfoy,
    {
        "Altin": 0.10,
        "EUR": -0.05,
        "TI1": 0.02
    }
)

print("Yeni toplam:", round(yeni_toplam,2))
yeni_portfoy.sort_values("deger_tl", ascending=False)
```

```python
#baz portfoy(degisklik yok)

baz = portfoy.copy()

#senaryo sonucu (uretilen yeni portfoy)

sen = yeni_portfoy.copy()

#karsilastirma tablosu 

kars = baz.merge(
    sen[["tur","varlik","deger_tl","agirlik_%"]],
    on=["tur","varlik"],
    suffixes=("_baz","_sen")
)

kars["fark_tl"] = (kars["deger_tl_sen"] - kars["deger_tl_baz"]).round(2)
kars["fark_%"] = ((kars["fark_tl"] / kars["deger_tl_baz"]) * 100).round(2)

display(kars.sort_values("fark_tl", ascending=False))
```

```python
senaryolar = {
    "Risk-off": {"Altin":0.88, "EUR": 0.003, "USD": 0.003, "AFT": -0.05,"BTE": -0.08, "IJB": -0.08},
    "Risk-on":  {"Altin": -0.05, "EUR": -0.02, "USD": -0.02, "AFT": 0.08, "BTE": 0.12, "IJB": 0.12},
    "Kur şoku": {"USD": 0.15, "EUR": 0.12},
    "Altın şoku": {"Altin": 0.15, "Gumus": 0.10},
    "Faiz düşüş": {"TI1": 0.01, "IST": 0.02, "IPV": 0.03},
}

sonuclar = []
for ad, deg in senaryolar.items():
    df_s, top = senaryo(portfoy,deg)
    sonuclar.append([ad, round(top,2)])

df_sonuclar = pd.DataFrame(sonuclar, columns=["Senaryo", "Toplam_TL"])
display(df_sonuclar.sort_values("Toplam_TL", ascending=False))
```

```python
import pandas as pd
from datetime import datetime
import os
import traceback

CSV_DOSYA = "portfoy.csv"
EXCEL_DOSYA = "portfoy_rapor.xlsx"

# ---- Manuel fiyatlar / kurlar (şimdilik) ----
ALTIN_TL_GR = 6000.0
GUMUS_TL_GR = 98.0
USDTRY = 42.9372
EURTRY = 50.4614

fiyat_map = {
    "Altin": ALTIN_TL_GR,
    "Gumus": GUMUS_TL_GR,
    "USD": USDTRY,
    "EUR": EURTRY,
}

def normalize_cols(df):
    df = df.copy()
    df.columns = [c.strip().lower() for c in df.columns]
    return df

def build_df_detay(df_raw):
    df = normalize_cols(df_raw)

    # Zorunlu kolon kontrolü
    required = {"varlik", "tur", "miktar"}
    missing = required - set(df.columns)
    if missing:
        raise ValueError(f"CSV'de zorunlu kolon(lar) eksik: {sorted(missing)}. "
                         f"Mevcut kolonlar: {df.columns.tolist()}")

    # Tip dönüşümleri
    df["miktar"] = pd.to_numeric(df["miktar"], errors="coerce").fillna(0)

    # Birim fiyat map (maden/döviz için)
    df["birim_fiyat_tl"] = df["varlik"].map(fiyat_map)

    # deger_tl oluştur (fonlarda tutar_tl veya deger_tl bekliyoruz)
    if "deger_tl" not in df.columns:
        if "tutar_tl" in df.columns:
            df["deger_tl"] = pd.to_numeric(df["tutar_tl"], errors="coerce").fillna(0.0)
        else:
            df["deger_tl"] = 0.0
    else:
        df["deger_tl"] = pd.to_numeric(df["deger_tl"], errors="coerce").fillna(0.0)

    # maden/döviz için hesap (fon değerini ezmeden)
    mask_md = df["tur"].isin(["maden", "doviz"])
    df.loc[mask_md, "deger_tl"] = df.loc[mask_md, "miktar"] * df.loc[mask_md, "birim_fiyat_tl"]

    toplam = float(df["deger_tl"].sum())
    if toplam <= 0:
        raise ValueError("Toplam portföy TL değeri 0 veya negatif görünüyor. "
                         "CSV'de deger_tl/tutar_tl yok ya da miktarlar 0.")

    df["agirlik_%"] = (df["deger_tl"] / toplam * 100).round(2)

    # Fonlar 0 mı? (sessiz kalmasın)
    uyarilar = []
    if "fon" in set(df["tur"].astype(str).str.lower()):
        fon_mask = df["tur"].astype(str).str.lower().eq("fon")
        fon_toplam = float(df.loc[fon_mask, "deger_tl"].sum())
        if fon_toplam == 0:
            uyarilar.append("Fonların TL değeri 0 görünüyor. CSV'ye fonlar için 'tutar_tl' (toplam TL) veya 'deger_tl' eklemelisin.")

    return df, toplam, uyarilar

# ---- Ana akış: hata olsa bile Excel'e en az 1 sheet yaz ----
try:
    if not os.path.exists(CSV_DOSYA):
        raise FileNotFoundError(f"{CSV_DOSYA} bulunamadı. Dosya bu notebook ile aynı klasörde olmalı.")

    df_raw = pd.read_csv(CSV_DOSYA)
    df_detay, toplam, uyarilar = build_df_detay(df_raw)

    with pd.ExcelWriter(EXCEL_DOSYA, engine="openpyxl", mode="w") as writer:
        df_detay.to_excel(writer, sheet_name="01_Portfoy_Detay", index=False)

        tur_ozet = df_detay.groupby("tur", dropna=False)["deger_tl"].sum().reset_index()
        tur_ozet.to_excel(writer, sheet_name="02_Tur_Ozet", index=False)

        kons = df_detay.sort_values("agirlik_%", ascending=False).head(10)
        kons.to_excel(writer, sheet_name="03_Konsantrasyon", index=False)

        snapshot = pd.DataFrame([{
            "Tarih": datetime.now().strftime("%Y-%m-%d %H:%M"),
            "Toplam_Portfoy_TL": round(toplam, 2),
            "Top1_%": float(df_detay.sort_values("agirlik_%", ascending=False).head(1)["agirlik_%"].values[0]),
            "Top3_%": float(df_detay.sort_values("agirlik_%", ascending=False).head(3)["agirlik_%"].sum()),
        }])
        snapshot.to_excel(writer, sheet_name="04_Snapshot", index=False)

        if uyarilar:
            pd.DataFrame({"Uyarilar": uyarilar}).to_excel(writer, sheet_name="05_Uyarilar", index=False)

    print("✅ df_detay hazır ve Excel yazıldı:", EXCEL_DOSYA)
    print("Toplam Portföy (TL):", round(toplam, 2))
    display(df_detay.sort_values("deger_tl", ascending=False).head(8))

except Exception as e:
    # Boş workbook hatası yememek için hata raporunu Excel'e basalım
    err_text = traceback.format_exc()

    with pd.ExcelWriter(EXCEL_DOSYA, engine="openpyxl", mode="w") as writer:
        pd.DataFrame({
            "Hata": [str(e)],
            "Detay": [err_text],
            "Not": ["CSV kolonlarını kontrol et veya df_detay üretim hücresini önce çalıştır."]
        }).to_excel(writer, sheet_name="00_ERROR", index=False)

    print("❌ Hata oluştu ama Excel'e '00_ERROR' sayfası yazıldı:", EXCEL_DOSYA)
    print(str(e))
```

```python
import pandas as pd

# 1) df_fon hazır değilse burada üret (senin verdiğin tablo)
fon_data = [
    ["TI1", "Fon - Para Piyasasi", 243150, "TL"],
    ["IST", "Fon - Kısa Vadeli Borç", 83339, "TL"],
    ["IPV", "Fon - Eurobond", 40695, "TL"],
    ["TTA", "Fon - Altın", 30862, "TL"],
    ["AFT", "Fon - Teknoloji", 13289, "TL"],
    ["TMG", "Fon - Yabancı Hisse", 10406, "TL"],
    ["NST", "Fon - İnşaat", 9247, "TL"],
]
df_fon = pd.DataFrame(fon_data, columns=["kod", "kategori", "tutar_tl", "para_birimi"])

# 2) df_detay mevcut olmalı (senin çalıştırdığın akıştan geliyor)
# df_detay yoksa önce "kurtarma kiti" hücreni çalıştır.
if "df_detay" not in globals():
    raise NameError("df_detay yok. Önce df_detay üreten hücreyi çalıştır.")

# 3) Kolon adlarını normalize et (güvenli olsun)
df_detay.columns = [c.strip().lower() for c in df_detay.columns]
df_fon.columns = [c.strip().lower() for c in df_fon.columns]

# 4) Fonları df_detay'a bağla (varlik = fon kodu)
# Sadece fon satırlarına tutar_tl enjekte edeceğiz
df_detay = df_detay.merge(
    df_fon[["kod", "tutar_tl"]],
    how="left",
    left_on="varlik",
    right_on="kod"
)

# 5) Fonlarda deger_tl 0 ise tutar_tl ile doldur
fon_mask = df_detay["tur"].astype(str).str.lower().eq("fon")
df_detay.loc[fon_mask & (df_detay["deger_tl"] == 0), "deger_tl"] = df_detay.loc[fon_mask, "tutar_tl"].fillna(0.0)

# 6) Temizlik: merge'den gelen kod kolonunu kaldır
df_detay = df_detay.drop(columns=["kod"], errors="ignore")

# 7) Ağırlıkları yeniden hesapla
toplam = float(df_detay["deger_tl"].sum())
df_detay["agirlik_%"] = (df_detay["deger_tl"] / toplam * 100).round(2)

print("✅ Fon TL değerleri df_detay'a işlendi.")
print("Toplam Portföy (TL):", round(toplam, 2))

display(df_detay.sort_values("deger_tl", ascending=False).head(12))
```
