```python
sayilar = [12, 7, 18, 5, 21, 30, 4]
```

```python
def ciftleri_bul(liste):
    return [x for x in liste if x % 2 == 0]

def tekleri_bul(liste):
    return [x for x in liste if x %2 != 0]

ciftler = ciftleri_bul(sayilar)
tekler = tekleri_bul(sayilar)

print("Ciftler", ciftler)
print("Tekler", tekler)
```

```python
if len(ciftler) > len(tekler):
    print("cift sayi agirlikli bir liste")
else:
    print("Tek sayi agirlikli bir liste")
```

```python
import pandas as pd 

df = pd.DataFrame({"sayi":sayilar})
df
```

```python
df["cift_mi"] = df["sayi"] % 2 == 0
df["tek_mi"] =df["sayi"] % 2 != 0
df
```

```python
ciftler_pd = df.loc[df["cift_mi"], "sayi"].tolist()
tekler_pd = df.loc[df["tek_mi"], "sayi"].tolist()

print("ciftler (pandas):", ciftler_pd)
print("tekler (pandas):", tekler_pd)

if len(ciftler_pd) > len(tekler_pd):
    print("pandas: cift sayi agirlikli")
else:
    print("pandas: tek sayi agirlikli")
```
