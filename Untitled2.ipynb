```python
import pandas as pd

def analiz_ve_raporla(giris_dosyasi, csv_cikis="rapor.csv", txt_cikis="ozet.txt"):
    with open(giris_dosyasi, "r", encoding="utf-8") as f:
        satir = f.read().strip()

    sayilar = [int(x) for x in satir.split(",")]

    df = pd.DataFrame({"sayi": sayilar})
    df["cift_mi"] = df["sayi"] % 2 == 0
    df["tek_mi"] = ~df["cift_mi"]

    # dosyaya yaz
    df.to_csv(csv_cikis, index=False)

    sayim = df["cift_mi"].value_counts().to_dict()
    ciftler = df.loc[df["cift_mi"], "sayi"].tolist()
    tekler  = df.loc[df["tek_mi"], "sayi"].tolist()

    with open(txt_cikis, "w", encoding="utf-8") as f:
        f.write(f"Giris: {giris_dosyasi}\n")
        f.write(f"Sayilar: {sayilar}\n")
        f.write(f"Ciftler: {ciftler}\n")
        f.write(f"Tekler: {tekler}\n")
        f.write(f"Sayim: {sayim}\n")

    return df, {"sayilar": sayilar, "ciftler": ciftler, "tekler": tekler, "sayim": sayim}

df, sonuc = analiz_ve_raporla("sayilar.txt", "rapor_sayilar.csv", "ozet_rapor.txt")
df
```

```python
import random

def sayi_tahmin_oyunu_v2():
    hedef = random.randint(1, 100)
    hak = 7
    onceki_fark = None

    print("1-100 arasında bir sayı tuttum.")
    print("7 hakkın var. Başlayalım.")

    while hak > 0:
        giris = input(f"Tahminin (kalan hak: {hak}): ")

        if not giris.isdigit():
            print("Sadece sayı gir.")
            continue

        tahmin = int(giris)

        if tahmin < 1 or tahmin > 100:
            print("1 ile 100 arasında gir.")
            continue

        fark = abs(hedef - tahmin)

        if tahmin == hedef:
            print(f"Bildin! Sayı: {hedef}. Kalan hak: {hak}. Skor: {hak * 10}")
            return

        if onceki_fark is None:
            print("İlk atış. Devam.")
        else:
            if fark < onceki_fark:
                print("Sıcak! Yaklaşıyorsun.")
            elif fark > onceki_fark:
                print("Soğuk. Uzaklaşıyorsun.")
            else:
                print("Aynı mesafe. İlginç.")

        if tahmin < hedef:
            print("İpucu: Daha büyük.")
        else:
            print("İpucu: Daha küçük.")

        onceki_fark = fark
        hak -= 1

    print(f"Oyun bitti. Sayı: {hedef}. Skor: 0")

sayi_tahmin_oyunu_v2()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def rapor_paketi_uret(txt_dosya, csv_cikis="rapor.csv", ozet_cikis="ozet.txt"):
    # 1) oku
    with open(txt_dosya, "r", encoding="utf-8") as f:
        satir = f.read().strip()

    # 2) parse
    sayilar = [int(x) for x in satir.split(",") if x.strip() != ""]

    # 3) df
    df = pd.DataFrame({"sayi": sayilar})

    # 4) ozet
    cift_sayisi = int((df["sayi"] % 2 == 0).sum())
    tek_sayisi  = int((df["sayi"] % 2 != 0).sum())
    ort = float(df["sayi"].mean())
    mn  = int(df["sayi"].min())
    mx  = int(df["sayi"].max())

    # 5) dosyaya yaz
    df.to_csv(csv_cikis, index=False)

    with open(ozet_cikis, "w", encoding="utf-8") as f:
        f.write(f"Girdi: {txt_dosya}\n")
        f.write(f"Toplam adet: {len(df)}\n")
        f.write(f"Çift: {cift_sayisi} | Tek: {tek_sayisi}\n")
        f.write(f"Min: {mn} | Max: {mx} | Ortalama: {ort:.2f}\n")

    # 6) grafikler
    plt.figure()
    plt.hist(df["sayi"])
    plt.title("Sayı Dağılımı")
    plt.xlabel("Sayı")
    plt.ylabel("Frekans")
    plt.tight_layout()
    plt.savefig("sayi_hist.png")
    plt.close()

    sayim = (df["sayi"] % 2 == 0).value_counts()
    plt.figure()
    plt.bar(["Tek", "Çift"], [sayim.get(False, 0), sayim.get(True, 0)])
    plt.title("Tek-Çift Dağılımı")
    plt.xlabel("Tip")
    plt.ylabel("Adet")
    plt.tight_layout()
    plt.savefig("tek_cift_bar.png")
    plt.close()

    return df

df = rapor_paketi_uret("sayilar.txt", "rapor_sayilar.csv", "ozet_rapor.txt")
df
```
