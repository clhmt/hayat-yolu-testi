```python
import pandas as pd

df = pd.read_csv("harcama_1.csv")
df
```

```python
df.groupby("kategori")["tutar"].sum()
```

```python
kategori_toplam = (
    df.groupby("kategori")["tutar"]
    .sum()
    .reset_index()
    .sort_values(by="tutar", ascending=False)
)

kategori_toplam
```

```python
import matplotlib.pyplot as plt

plt.bar(kategori_toplam["kategori"], kategori_toplam["tutar"])



import matplotlib.pyplot as plt

plt.bar(kategori_toplam["kategori"], kategori_toplam["tutar"])
plt.title("kategori Bazli Toplam Harcama")
plt.xlabel("kategori")
plt.ylabel("Toplam Tutar")
plt.show()
```

```python
kategori_toplam.to_csv("kategori_ozet.csv", index=False)
```

```python
import pandas as pd

df = pd.read_csv("harcama_1.csv")
df
```

```python
df["kategori"] = df["kategori"].str.strip()

kategori_toplam = (
    df.groupby("kategori")["tutar"]
    .sum()
    .reset_index()
    .sort_values(by="tutar", ascending=False)
)

kategori_toplam
            
```

```python
kategori_toplam.to_csv("kategori_ozet.csv", index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.bar(kategori_toplam["kategori"], kategori_toplam["tutar"])
plt.title("kategori bazli harcama")
plt.xlabel("kategori")
plt.ylabel("toplam tutar")



plt.savefig("kategori_harcama.png")
plt.show()
```
